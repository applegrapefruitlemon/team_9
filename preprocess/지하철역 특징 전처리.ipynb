### 2) 불필요한 열 제거하기

```python
del_cols = subway[['하행순번', '상행순번', '방면구분', '구간거리', '구간시간', '정차시간',  '목적지코드',
       '열차정차여부', '상행열차정차여부', '하행열차정차여부']]
subway = subway.drop(del_cols, axis = 1)
```

```python
subway.head()
```

### 3) 호선명 매핑 

```python
호선참조 = pd.read_excel("실시간도착_역정보(20250513).xlsx")
호선참조.head()
```

```python
# 열 이름을 같게 맞추기
호선참조.rename(columns={'STATN_ID': '지하철역ID'}, inplace=True)
호선참조.head()
```

```python
subway['호선이름'] = subway['지하철역ID'].map(
    dict(zip(호선참조['지하철역ID'], 호선참조['호선이름']))
)
```
