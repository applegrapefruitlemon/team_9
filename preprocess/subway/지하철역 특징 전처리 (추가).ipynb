```python
def classify_wind_chill(temp):
    # 겨울 기준
    if temp < -15.4:
        return '위험(겨울)'
    elif -15.4 <= temp < -10.5:
        return '경고(겨울)'
    elif -10.5 <= temp < -3.2:
        return '주의(겨울)'
    elif -3.2 <= temp < 0:
        return '관심(겨울)'
    # 여름 기준
    elif 31 <= temp < 33:
        return '관심(여름)'
    elif 33 <= temp < 35:
        return '주의(여름)'
    elif 35 <= temp < 38:
        return '경고(여름)'
    elif temp >= 38:
        return '위험(여름)'
    else:
        return '일반'

# 적용
df['체감온도구간'] = df['체감온도'].apply(classify_wind_chill)

print(df)
```

### 6) 위경도 매핑

```python
import pandas as pd
df = pd.read_csv("merged_혼잡도_역특징_updated.csv")
df
```

```python
# 역별로 변경할 값을 딕셔너리로 정리
update_dict = {
    "대방": "O",
    "신림": "O",
    "올림픽공원": "O",
    "양평": "X",
    "보라매": "O",
    "구리": "O",
    "별내": "O",
    "샛강": "O",
    "달월": "X",
    "월곶": "X",
    "소래포구": "X",
    "인천논현": "X",
    "호구포": "X",
    "남동인더스파크": "X",
    "연수": "X",
    "송도": "X",
    "인하대": "X",
    "숭의": "X",
    "신포": "X"
}

# 역명 열이 '역명', 환승 여부 열이 '환승역여부'라고 가정
df["환승역여부"] = df.apply(
    lambda row: update_dict[row["역명"]] if row["역명"] in update_dict else row["환승역여부"],
    axis=1
)
```

```python
import math
NX = 149            ## X축 격자점 수
NY = 253            ## Y축 격자점 수

Re = 6371.00877     ##  지도반경
grid = 5.0          ##  격자간격 (km)
slat1 = 30.0        ##  표준위도 1
slat2 = 60.0        ##  표준위도 2
olon = 126.0        ##  기준점 경도
olat = 38.0         ##  기준점 위도
xo = 210 / grid     ##  기준점 X좌표
yo = 675 / grid     ##  기준점 Y좌표
first = 0

if first == 0 :
    PI = math.asin(1.0) * 2.0
    DEGRAD = PI/ 180.0
    RADDEG = 180.0 / PI


    re = Re / grid
    slat1 = slat1 * DEGRAD
    slat2 = slat2 * DEGRAD
    olon = olon * DEGRAD
    olat = olat * DEGRAD

    sn = math.tan(PI * 0.25 + slat2 * 0.5) / math.tan(PI * 0.25 + slat1 * 0.5)
    sn = math.log(math.cos(slat1) / math.cos(slat2)) / math.log(sn)
    sf = math.tan(PI * 0.25 + slat1 * 0.5)
    sf = math.pow(sf, sn) * math.cos(slat1) / sn
    ro = math.tan(PI * 0.25 + olat * 0.5)
    ro = re * sf / math.pow(ro, sn)
    first = 1

def mapToGrid(lat, lon, code = 0 ):
    ra = math.tan(PI * 0.25 + lat * DEGRAD * 0.5)
    ra = re * sf / pow(ra, sn)
    theta = lon * DEGRAD - olon
    if theta > PI :
        theta -= 2.0 * PI
    if theta < -PI :
        theta += 2.0 * PI
    theta *= sn
    x = (ra * math.sin(theta)) + xo
    y = (ro - ra * math.cos(theta)) + yo
    x = int(x + 1.5)
    y = int(y + 1.5)
    return x, y

def gridToMap(x, y, code = 1):
    x = x - 1
    y = y - 1
    xn = x - xo
    yn = ro - y + yo
    ra = math.sqrt(xn * xn + yn * yn)
    if sn < 0.0 :
        ra = -ra
    alat = math.pow((re * sf / ra), (1.0 / sn))
    alat = 2.0 * math.atan(alat) - PI * 0.5
    if math.fabs(xn) <= 0.0 :
        theta = 0.0
    else :
        if math.fabs(yn) <= 0.0 :
            theta = PI * 0.5
            if xn < 0.0 :
                theta = -theta
        else :
            theta = math.atan2(xn, yn)
    alon = theta / sn + olon
    lat = alat * RADDEG
    lon = alon * RADDEG

    return lat, lon


chunksize = 1000
results = []

for start in range(0, len(df), chunksize):
    end = start + chunksize
    chunk = df.iloc[start:end].copy()

    chunk[['위도', '경도']] = chunk.apply(
        lambda row: pd.Series(gridToMap(row['nx'], row['ny'])), axis=1
    )
    
    results.append(chunk)

df_converted = pd.concat(results, ignore_index=True)

print(df_converted)
```

```python
# (지하철역ID, 역명) 조합별로 첫 번째 행만 뽑아서 고유값 DataFrame 만들기
unique_stations = df_converted.drop_duplicates(subset=['지하철역ID', '역명'])

# 딕셔너리 생성
lat_dict = unique_stations.set_index(['지하철역ID', '역명'])['위도'].to_dict()
lon_dict = unique_stations.set_index(['지하철역ID', '역명'])['경도'].to_dict()

# 매핑 적용
df_converted['이전_위도'] = df_converted.apply(
    lambda row: lat_dict.get((row['이전지하철역ID'], row['이전지하철역명'])), axis=1
)
df_converted['이전_경도'] = df_converted.apply(
    lambda row: lon_dict.get((row['이전지하철역ID'], row['이전지하철역명'])), axis=1
)
df_converted['다음_위도'] = df_converted.apply(
    lambda row: lat_dict.get((row['다음지하철역ID'], row['다음지하철역명'])), axis=1
)
df_converted['다음_경도'] = df_converted.apply(
    lambda row: lon_dict.get((row['다음지하철역ID'], row['다음지하철역명'])), axis=1
)
print(df_converted)
```

```python
import pandas as pd
# csv파일 불러오기
csv = pd.read_csv('merged_테블로진행용_updated_구간.csv',encoding='cp949')
print(csv.head())

# 데이터프레임 주소값 추출
address= csv['기본주소']
print(address.head())


# 주소 데이터 깔끔하게 다듬기
for i in range(len(address)):
    a = address[i].split(' ')
    address[i] = " ".join(a[0:4])
print(address)

####### 도로명주소 위도 경도 값으로 바꿔주기 ########
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]




#####주소를 위,경도 값으로 변환하기 #####
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])

#####Dataframe만들기######
address_df = pd.DataFrame({'카페이름': csv['사업장명'],'상세주소':csv['소재지도로명주소'],'주소':address,'위도':latitude,'경도':longitude})

#df저장
address_df.to_csv('jungu_incheon_coffeeshop.csv')
```
