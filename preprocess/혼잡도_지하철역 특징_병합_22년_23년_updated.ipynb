```python
import pandas as pd
```

```python
혼잡도 = pd.read_csv("혼잡도_2223_updated_속성변경.csv")
역특징 = pd.read_csv("지하철_특징별_혼잡도역명반영.csv", encoding = 'euc-kr')
display(혼잡도, 역특징)
```

```python
역특징.rename(columns= {"지하철호선명":"호선"}, inplace=True)
역특징.head()
```

```python
import pandas as pd

# '역명'과 '호선' 열을 기준으로 병합
merged_혼잡도_역특징 = pd.merge(혼잡도, 역특징, on=['역명', '호선'], how='inner')  # inner, left, right, outer 선택 가능

# 결과 저장 (필요시)
merged_혼잡도_역특징.to_csv('merged_혼잡도_역특징_updated.csv', index=False)
```
